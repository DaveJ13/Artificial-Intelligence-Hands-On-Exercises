```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# A feedforward neural network is the simplest artificial neural network containing multiple nodes arranged
# in multiple layers. Adjacent layer nodes have connections or edges where all connections are weighted
# A feedforward neural network can contain two kinds of nodes
# Monolayer: The simplest feedforward neural network that doesn't contain any hidden layers
# Multi-layer perceptron: (MLP) contains at least one hidden layer(except for one input and one output layer)
```

```python
# Multi-layer perceptron
# A multi-layer perceptron(MLP) is a deep, artificial neural network
# It is composed of more than one perceptron
# An MLP is comprised of:
  # 1. An input layer to receive the signal
  # 2. An output layer that makes a decision or prediction about the input
  # 3. An arbitrary number of hidden layers
# Each node, apart from the input nodes, has a nonlinear activation function
# An MLP uses backpropogation as a supervised learning technique
```

```python
# MLP is widely used for solving problems that require supervised learning and research into computational
# neuroscience and parallel distributed processing. Such applications include speech recognition,
# image recognition, and machine translation
```

```python
# Feedforward neural network 
x1 = np.random.randint(3,12,10)  # 10 random values in the range of 3-12
x2 = np.random.randint(9,18,10)  # 10 random values in the range of 9-18
x3 = np.random.randint(12,20,10)  # 10 random values in the range of 12-20
y = x1*7 + 5*x2 + 4*x3
```

```python
x1,x2,x3
```

```python
SHAPE = x1.shape[0]
SHAPE
```

```python
y   # This is a numpy array
```

```python
y = y.reshape(SHAPE,1)   # reshape(# of rows, # of columns)
y
```

```python
X = np.array([x1,x2,x3])
X
```

```python
X = X.T
X
```

```python
lr = .0001
w = np.array([np.random.randn(1),
              np.random.randn(1),
              np.random.randn(1)])
w
```

```python
error_list = []
for i in range(50):   # Number of epochs is 50 (number of back propogations)
    y_pred = np.matmul(X, w)   # Using matrix multiplication to find the initial predicted values
    # X is the input matrix and w is the weight values
    error = (y - y_pred)**2
    dEw1 = 2*np.matmul((y - y_pred).T, (-X[:,0].reshape(SHAPE, 1)))  # change in error with respect to weight 1
    dEw2 = 2*np.matmul((y - y_pred).T, (-X[:,1].reshape(SHAPE, 1)))  # change in error with respect to weight 2
    dEw3 = 2*np.matmul((y - y_pred).T, (-X[:,2].reshape(SHAPE, 1)))  # change in error with respect to weight 3
    
    w[0][0] = w[0][0] - lr * dEw1
    w[1][0] = w[1][0] - lr * dEw2
    w[2][0] = w[2][0] - lr * dEw3
    
    error_list.append(error.mean())
    print("Value of w1 - ",w[0][0], "Value of w2 - ",w[1][0], "Value of w3 - ",w[2][0], error)
```

```python
plt.plot(error_list)
```
